```python
import seaborn as sns
import matplotlib.pyplot as plt
from tabulate import tabulate

# configure matplotlib to output inline
%matplotlib inline
```

```python
# victims of rape DataFrame
rape_victims = pd.read_csv('/kaggle/input/crime-in-india/20_Victims_of_rape.csv')

# let's see how the data is structured
rape_victims.head()
```

**What is the overall trend in the number of reported rape cases against women over the years? Visualize this trend using a line chart.**

```python
# 1. Trend Analysis
trend_data = rape_victims.groupby('Year')['Rape_Cases_Reported'].sum()
plt.plot(trend_data.index, trend_data.values)
plt.title('Trend Analysis of Reported Rape Cases')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
```

The number of reported rape cases increased from about 34,000 in 2002 to about 42,000 in 2010.

**Which age group has the highest number of reported rape cases? Create a bar chart to compare the distribution of cases across different age groups.**

```python
# 2. Age Group Analysis
age_group_data = rape_victims.groupby('Subgroup')[['Victims_Above_50_Yrs', 'Victims_Between_10-14_Yrs', 'Victims_Between_14-18_Yrs', 'Victims_Between_18-30_Yrs', 'Victims_Between_30-50_Yrs', 'Victims_Upto_10_Yrs']].sum()
age_group_data.plot(kind='bar', stacked=True)
plt.title('Age Group Analysis of Reported Rape Cases')
plt.xlabel('Subgroup')
plt.ylabel('Number of Cases')
plt.show()
```

**How do the numbers vary across different states or regions? Create a geographical heatmap to visualize the regional distribution of reported cases.**

```python
# 3. Regional Comparison
plt.figure(figsize=(10, 6))
sns.heatmap(rape_victims.pivot_table(values='Rape_Cases_Reported', index='Area_Name', columns='Year', aggfunc='sum'), cmap='YlGnBu')
plt.title('Regional Comparison of Reported Rape Cases Over Years')
plt.xlabel('Year')
plt.ylabel('Area Name')
plt.show()
```

**Analyze the distribution of cases based on subgroups such as "Total Rape Victims," "Victims of Incest Rape," and "Victims of Other Rape." Visualize this information using a stacked bar chart.**

```python
# 4. Subgroup Analysis
subgroup_data = rape_victims.pivot_table(values='Rape_Cases_Reported', index='Year', columns='Subgroup', aggfunc='sum')
subgroup_data.plot(kind='bar', stacked=True)
plt.title('Subgroup Analysis of Reported Rape Cases Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
```

**Are there any specific years with a significant increase or decrease in reported cases? Visualize the yearly variation using a bar chart or line chart.**

```python
# 5. Yearly Variation
yearly_variation_data = rape_victims.groupby('Year')['Rape_Cases_Reported'].sum()
yearly_variation_data.plot(kind='bar')
plt.title('Yearly Variation of Reported Rape Cases')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
```

**Explore the demographics of victims, such as their age and gender. Create a pie chart or donut chart to represent the percentage distribution.**

```python
# 6. Victim Demographics
victim_demographics_data = rape_victims[['Victims_Above_50_Yrs', 'Victims_Between_10-14_Yrs', 'Victims_Between_14-18_Yrs', 'Victims_Between_18-30_Yrs', 'Victims_Between_30-50_Yrs', 'Victims_Upto_10_Yrs']].sum()
victim_demographics_data.plot(kind='pie', autopct='%1.1f%%')
plt.title('Demographics of Rape Victims')
plt.show()
```

**Investigate if there is any correlation between the age group of victims and the reported cases. Use a scatter plot to visualize this correlation.**

```python
# 7. Correlation Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Victims_Between_18-30_Yrs', y='Rape_Cases_Reported', data=rape_victims)
plt.title('Correlation Analysis between Age Group and Reported Rape Cases')
plt.xlabel('Victims Between 18-30 Yrs')
plt.ylabel('Number of Cases')
plt.show()
```

**How has the distribution of reported cases in different age groups changed over the years? Visualize this using a stacked area chart.**

```python
# 8. Age Group Comparison Over Time
age_group_comparison_data = rape_victims.groupby(['Year', 'Subgroup'])[['Victims_Between_18-30_Yrs', 'Victims_Between_30-50_Yrs']].sum().unstack()
age_group_comparison_data.plot(kind='area', stacked=False)
plt.title('Age Group Comparison of Reported Rape Cases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
```

**Compare the number of reported cases each year across different states. Use a grouped bar chart for this comparison.**

```python
# 9. Yearly Comparison Across States
state_comparison_data = rape_victims.pivot_table(values='Rape_Cases_Reported', index='Year', columns='Area_Name', aggfunc='sum')
state_comparison_data.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Yearly Comparison of Reported Rape Cases Across States')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
```

**Calculate and visualize the incidence rate of reported rape cases per 100,000 women in each state. Create a choropleth map to represent this information geographically.**

```python
# 10. Incidence Rate Calculation
rape_victims['Incidence_Rate'] = (rape_victims['Rape_Cases_Reported'] / rape_victims['Victims_of_Rape_Total']) * 100000
state_incidence_rate_data = rape_victims.pivot_table(values='Incidence_Rate', index='Year', columns='Area_Name', aggfunc='mean')
state_incidence_rate_data.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Incidence Rate of Reported Rape Cases Across States Over Years')
plt.xlabel('Year')
plt.ylabel('Incidence Rate (per 100,000 women)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

df_filtered = df[df['Subgroup'] == 'Total Rape Victims']

predictions = []
for state in df_filtered['Area_Name'].unique():
    temp = df_filtered[df_filtered['Area_Name'] == state].sort_values(by='Year')
    if len(temp) < 2:
        continue
    
    X = temp[['Year']]
    y = temp['Rape_Cases_Reported']

    model = LinearRegression()
    model.fit(X, y)
    
    next_year = pd.DataFrame({'Year': [temp['Year'].max() + 1]})
    pred = model.predict(next_year)[0]
    
    predictions.append({'Area_Name': state, 'Predicted_Cases_2025': int(pred)})

pred_df = pd.DataFrame(predictions).sort_values(by='Predicted_Cases_2025', ascending=False)
pred_df.head(10)
```
